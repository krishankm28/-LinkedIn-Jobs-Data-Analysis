```python
#converting into same unit
import numpy as np

def to_hours(value):
    if isinstance(value, str):
        value = value.lower().strip()
        try:
            num = float(value.split()[0])
        except:
            return np.nan

        if "minute" in value:
            return round(num / 60, 2)       # minutes → hours
        elif "hour" in value:
            return round(num, 2)            # already hours
        elif "day" in value:
            return round(num * 24, 2)       # days → hours
    return np.nan

df['posted_day_ago'] = df['posted_day_ago'].apply(to_hours)
```
